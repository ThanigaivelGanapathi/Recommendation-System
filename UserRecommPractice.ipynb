```python
import numpy as np
import pandas as pd
```

```python
r_cols = ['user_id','movie_id','rating','unix_timestamp']
dataset = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
dataset
```

```python
userCount  = dataset.user_id.unique().shape[0]
movieCount = dataset.movie_id.unique().shape[0]

print("Unique Users : ", userCount)
print("Unique Movies : ", movieCount)
```

```python
pivot_table = dataset.pivot_table(index = 'user_id',columns='movie_id',values='rating')
pivot_table
```

```python
pivot_table = pivot_table.replace(np.nan,0)
pivot_table
```

# Cosing Similarity

```python
from sklearn.metrics.pairwise import pairwise_distances

user_similarity = pairwise_distances(pivot_table, metric = 'cosine')
user_similarity
```

```python
pivot_array = pivot_table.to_numpy()
mean_user_rating = pivot_array.mean(axis=1)  # shape: (n_users,)
mean_user_rating = mean_user_rating[:, np.newaxis]  # reshape to (n_users, 1)
ratings_diff = pivot_array - mean_user_rating  # broadcasting works
numerator = user_similarity.dot(ratings_diff)  # shape: (n_users, n_items)
denominator = np.abs(user_similarity).sum(axis=1)[:, np.newaxis]  # shape: (n_users, 1)
pred = mean_user_rating + numerator / denominator  # shape: (n_users, n_items)
pred
```

```python
similar_user_movie_list = []

for user in similar_user:
    similar_user_movie_list.append(list(dataset[dataset['user_id'] == user]['movie_id']))

similar_user_movie_list
```
