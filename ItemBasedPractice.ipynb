```python
import numpy as np
import pandas as pd
```

```python
r_cols = ["user_id","movie_id","rating","unix_timestamp"]
dataset = pd.read_csv("ml-100k/u.data",sep = "\t", names = r_cols, encoding = "latin-1")
dataset
```

```python
n_users = dataset.user_id.unique().shape[0]
n_movies = dataset.movie_id.unique().shape[0]

print("Total User : ", n_users)
print("Total Movies : ", n_movies)
```

```python
#Creating Pivot Table

pivot_table = dataset.pivot_table(index = 'user_id', columns = 'movie_id', values = 'rating')
pivot_table
```

```python
pivot_table = pivot_table.replace(np.NaN,0)
pivot_table
```

```python
#Cosine Similarity

from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(pivot_table, metric='cosine')
item_similarity = pairwise_distances(pivot_table.T, metric='cosine')
```

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1).to_numpy()  # Convert to NumPy array
        ratings_diff = ratings.to_numpy() - mean_user_rating[:, np.newaxis]
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.abs(similarity).sum(axis=1)[:, np.newaxis]
    elif type == 'item':
       # pred = pivot_table.to_numpy().dot(similarity) / np.abs(similarity).sum(axis=1)[np.newaxis, :]
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python
user_prediction = predict(pivot_table, user_similarity, type='user')
item_prediction = predict(pivot_table, item_similarity, type='item')
```
